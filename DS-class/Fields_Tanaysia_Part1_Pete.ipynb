#### Type reasoning

- Student ID (int): Chosen as an integer because it is a unique numeric identifier with no decimal values  
- Student Name (str): Chosen as a string because it stores text data.  
- GPA (float): Chosen as a float because GPA includes decimal values.  
- Is_the_student_enrolled (bool): Chosen as a boolean because it represents a True/False condition.  
- Students_major (str): Chosen as a string because it stores the name of a major (text).  
- Credits_completed (int): Chosen as an integer because credits are counted as whole numbers.  
- Tuition_balance (float): Chosen as a float because monetary values can include decimals.  
- Is_on_honor_roll (bool): Chosen as a boolean because it represents a True/False status.
